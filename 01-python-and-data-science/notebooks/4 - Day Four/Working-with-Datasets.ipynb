We want to create a model that answers a question about some recorded results, based on the other information in the data (features/columns), that isn't easily understood just from looking at the data.

For example, what features of your life, explain your grades?

target: 
* grades

possible features:
* sleep
* attendance
* tardies
* absences
* study time
* missing assignments
* time on extra curricular activies

* commute times

## How can I select a dataset?

Examples of datasets that work well:
 1. [Most Streamed Spotify Songs 2024](https://www.kaggle.com/datasets/nelgiriyewithana/most-streamed-spotify-songs-2024)
    * Target: All Time Rank
 1. [Top 100 Youtube Channels in the World](https://www.kaggle.com/datasets/mayankanand2701/top-1000-youtube-channels-in-the-world)
    * Target: Subscribers


Examples of simpler datasets that can be used to make a point:
 1. [Fortune Top 1000 Companies by Revenue 2022](https://www.kaggle.com/datasets/surajjha101/fortune-top-1000-companies-by-revenue-2022)
    * Target: Revenue
    * Reason: A lot of the data readily explains the revenue, and the question of how they earned that revenue is not as interesting. However, you could use three features to explain the revenue.
 1. [Film Genre Statistics](https://www.kaggle.com/datasets/thedevastator/film-genre-statistics)
    * Target: Inflation-Adjusted Gross
    * Reason: We have a good amount of data to use features to explain how much money a film genre can make, but there are limited columns with data.

Examples of datasets that don't work (or are less useful for feature engineering):
 1. [best-selling-books](https://www.kaggle.com/datasets/drahulsingh/best-selling-books)
    * Reason: It just lists data, that we can't really use to explain something about a ranking/score/sales

Some suggestions for finding good datasets:
* Has 'all time rank' in the title, or something with 'rankings/scores'.
